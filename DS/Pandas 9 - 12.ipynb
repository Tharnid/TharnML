```python
import pandas as pd

movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]
```

```python
# logical operators and / or

movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
# or example

movies[(movies.duration >= 200) | (movies.genre == 'Drama')] # pipe symbol is or
```

```python
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')]
```

```python
# previous example works but is very wordy

movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```

```python
# Vid 12 String methods in Pandas

import pandas as pd

orders = pd.read_table('http://bit.ly/chiporders')

orders.head()
```

```python
# convert column heading to uppercase

orders.item_name.str.upper()
```

```python
# contains chickin

orders.item_name.str.contains('Chicken')
```

```python
# chain string methods together

orders.choice_description.str.replace('[', '')
```

```python
# replacing both brackets

orders.choice_description.str.replace('[', '').str.replace(']', '')
```

```python
# doing it with a regular expression

orders.choice_description.str.replace('[\[\]]', '')
```
