```python
# vid 19 - Selecting multiple rows and columns from a pandas DF

ufo = pd.read_csv('http://bit.ly/uforeports')

ufo.head(3) # first three rows
```

```python
# loc is more powerful

ufo.loc[0, :] # filtering rows and selecting columns by label rows, columns : means all columns
```

```python
# rows 1 2 3
ufo.loc[[0, 1, 2], :] # all columns
```

```python
# simpler way

ufo.loc[0:2, :] # inclusive on both sides
```

```python
ufo.loc[0:2] # assumes , : is there
```

```python
ufo.loc[:, ['City']]
```

```python
# multiple columns

ufo.loc[:, ['City', 'State']]
```

```python
# specify it as a range

ufo.loc[:, 'City':'State']
```

```python
# columns 0,2 through city state

ufo.loc[0:2, 'City':'State']
```

```python
# another method without loc

# dropping Time

ufo.head(3).drop('Time', axis=1)
```

```python
# loc with booleans

ufo[ufo.City=='Oakland']
```

```python
# same with loc..a bit more explicit

ufo.loc[ufo.City=='Oakland', :]
```

```python
ufo.loc[ufo.City=='Indianapolis', :]
```

```python
# just want state column

ufo.loc[ufo.City=='Indianapolis', 'State']
```

```python
# Slight differenc how this code works

# ufo[ufo.City=='Oakland'].State

# Chained indexing - problems under certain scenarios

# Compared to...

# ufo.loc[ufo.City=='Indianapolis', 'State']

# Safer for your code to execute

# iloc

ufo.iloc[:, [0, 3]]
```

```python
ufo.columns
```

```python
ufo.iloc[:, 0:4] # exclusive of second number....loc inclusive on both sides
```
